# 1. Bola

---


Bola adalah bangun ruang dengan bentuk bulat sempurna yang tersusun oleh tidak terhingga lingkaran yang mempunyai jari-jari dan pusat lingkaran yang sama.


image.png

```
phi=(22/7)
r=int(input("Kalau Jari-jarinya itu "))

luasPermukaan = int(4*phi*(r*r))
volume = int(4/3*phi*r*r*r)

print("Maka Luas Bolanya adalah", luasPermukaan)
print("dan Volume Bolanya adalah", volume)
```

# 2. Kerucut


---


Kerucut adalah bangun ruang yang memiliki 2 sisi yaitu sebuah lingkaran dan sebuah bidang lengkung.


image.png

```
phi=(22/7)
r=int(input("Kalau Jari-jarinya itu "))
t=int(input("terus Tingginya itu "))
s=int(input("dan Panjang Garis Pelukisnya itu "))

luasPermukaan = int((phi*r*r) + (phi*r*s))
volume = int(1/3*phi*r*r*t)

print("Maka Luas Kerucutnya adalah", luasPermukaan)
print("dan Volume Kerucutnya adalah", volume)
```

# 3. Tabung


---


Tabung adalah bangun ruang yang tersusun oleh 3 buah sisi yaitu 2 buah lingkaran yang mempunyai ukuran yang sama dan 1 segiempat yang mengelilingi kedua lingkaran tersebut.

image.png

```
phi=(22/7)
r=int(input("Kalau Jari-jarinya itu "))
t=int(input("dan Tingginya itu "))

luasPermukaan = int(2*phi*r*(r+t))
volume = int(phi*r*r*t)

print("Maka Luas Tabungnya adalah", luasPermukaan)
print("dan Volume Tabungnya adalah", volume)
```
